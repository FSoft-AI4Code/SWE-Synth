```python
def group_moatless_traj(df: pd.DataFrame):
    return (
        df.groupby(['exp_name', 'instance_id'])
        .agg({
            'messages': list,
        })
        .assign(
            traj_length=lambda x: x['messages'].apply(len)
        )
    )
```

```python
mutant_df = mutant_df.pipe(group_moatless_traj)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Compute value counts for all instances
strategy_counts = (
    df
    .assign(strategy=lambda x: x['synbugs_mutation_info'].apply(lambda y: y['strategy']))
    [['instance_id', 'strategy']]
    .value_counts('strategy')
)

# Compute value counts for solved instances
solved_strategy_counts = (
    df
    .assign(strategy=lambda x: x['synbugs_mutation_info'].apply(lambda y: y['strategy']))
    [['instance_id', 'strategy']]
    .query('instance_id in @solved_instance_ids')
    .value_counts('strategy')
)

# Align indices to ensure both data series have the same categories
all_strategies = strategy_counts.index.union(solved_strategy_counts.index)
strategy_counts = strategy_counts.reindex(all_strategies, fill_value=0)
solved_strategy_counts = solved_strategy_counts.reindex(all_strategies, fill_value=0)

# Compute percentages
total_instances = strategy_counts.sum()
total_solved_instances = solved_strategy_counts.sum()

strategy_percent = (strategy_counts / total_instances) * 100
solved_strategy_percent = (solved_strategy_counts / total_solved_instances) * 100

# Create the bar plot
fig, ax = plt.subplots(figsize=(10, 6))

y_positions = np.arange(len(all_strategies))  # Bar positions
bar_width = 0.4  # Width of bars

ax.barh(y_positions - bar_width / 2, strategy_counts, height=bar_width, label='All Instances (Count)')
ax.barh(y_positions + bar_width / 2, solved_strategy_counts, height=bar_width, label='Solved Instances (Count)')

# Annotate bars with percentage values
for i, (count, percent) in enumerate(zip(strategy_counts, strategy_percent)):
    ax.text(count + 50, i - bar_width / 2, f'{percent:.1f}%', va='center')

for i, (count, percent) in enumerate(zip(solved_strategy_counts, solved_strategy_percent)):
    ax.text(count + 50, i + bar_width / 2, f'{percent:.1f}%', va='center')

# Labels and title
ax.set_yticks(y_positions)
ax.set_yticklabels(all_strategies)
ax.set_xlabel('Count')
ax.set_ylabel('Strategy')
ax.set_title('Strategy Distribution (Counts & Percentages)')
ax.legend()

plt.grid()
plt.show()

# class strategy -> kho hon

# chot theo generated mutants
```

```python
import zstandard as zstd
import json
from pathlib import Path

OUTPUT_DIR = Path('/datadrive5/user3/user47/phase-5/ablation/strategy/')

def each_strategy(df: pd.DataFrame):
    strategy_name = df['strategy'].iloc[0]
    data = (
        df
        .to_dict(orient='records')
    )
    file_name = (OUTPUT_DIR / strategy_name).with_suffix('.jsonl.zst')
    with zstd.open(file_name, 'wt') as f:
        output_str = '\n'.join(json.dumps(o) for o in data)
        f.write(output_str)
    
    print(f'Finished writing to {file_name} with {len(data)} instances')

(
    out_df
    .reset_index()
    .drop(columns=['traj_length', 'level_1'])
    [['strategy', 'exp_name', 'instance_id', 'messages']]
    .explode('messages')
    .groupby('strategy')
    .apply(
        each_strategy,
        include_groups=True
    )
)
```
